# Data Cleaning

##### A code that is applicable for other columns were written once and used for those other columns, therefore running

```python
df.drop(labels=['Photo','Flag','Club Logo'], axis=1, inplace=True)  # getting rid of unwanted columns
```

```python
df['Value'].replace(to_replace='\€',value = '',inplace=True, regex=True)
```

```python
df['Value'].replace(to_replace='M',value = '',inplace=True, regex=True)
```

```python
df['Value'].replace(to_replace='K',value = '',inplace=True, regex=True)
```

```python
df['Value'] = df['Value'].astype(float)
```

```python
df.Value
```

```python
df['Joined'] = pd.to_datetime(df['Joined'])
```

```python
df['Joined'] = df['Joined'].apply(lambda x: x.year)
```

```python
df.Joined.isna().value_counts()
```

```python
df.Joined = df.Joined.astype(int)
```

```python
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'])
```

```python
df['Contract Valid Until'].describe()
```

```python
df['Height'].replace(to_replace="\'",value = '.', regex=True,inplace=True)
```

```python
df['Height'] = df['Height'].astype(float)
```

```python
df.Height.describe()
```

```python
def remove_suffix(x):
    a = x[0:3]
    return a
```

```python
df['Weight'] = df['Weight'].apply(remove_suffix)
```

```python
df['Weight'] = df['Weight'].astype(str)
```

```python
df['Weight']
```

```python
df['Weight'] = df['Weight'].astype(float)
```

```python
df['Release Clause'].replace(to_replace='\€',value='', regex=True, inplace=True)
```

```python
df['Release Clause'] = df['Release Clause'].astype(float)
```

##### Filling all missing values by mean imputation function.

```python
def myfillna(series):
    if series.dtype is pd.np.dtype(float):
        return series.fillna(series.mean())
    else:
        return series
```

```python
df = df.apply(myfillna)
```

```python
df.to_csv('mini_project.csv')
```

# Exploratory Analysis

```python
df1 = pd.read_csv('mini_project.csv')
```

#### 1.Plot the distribution of Overall rating for all players. 

```python
plt.figure(figsize=(10,8))
sns.distplot(df1.Overall,label='Overall Ratings',color='cyan')
plt.grid(axis='both', color='pink', linestyle='--', linewidth = 0.3)
plt.axvline(df1.Overall.mean(), color = 'red', label='Mean')
plt.axvline(df1.Overall.median(), color = 'hotpink', label='Median')
plt.axvline(df1.Overall.quantile(0.25), color = 'black', label = '1st quantile')
plt.axvline(df1.Overall.quantile(0.75), color = 'maroon', label = '3rd quantile')
plt.title('Distribution plot of Overall player ratings')
plt.legend()
plt.savefig('1_1_Overall_distribution')
```

```python
df1.Overall.kurt()
```

```python
df1.Overall.skew()
```

##### Observation

- Overall rating feature has a bell-shaped curve, therefore it's distribution is normal and it's peak follows the mesokurtic shape

#### 2.	Generate pair plots for the following variables:
